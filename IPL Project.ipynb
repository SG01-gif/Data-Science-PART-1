```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

```

```python
df=pd.read_csv('IPL.csv')
df.head()
```

```python
df.info()
```

```python
df.shape[0] # rows - 74
df.shape[1] # columns - 20
```

```python
df.isnull().sum() # no null values in the dataset

# 0 - not any null val
# 1 - null val present
```

```python
match_wins=df['match_winner'].value_counts()
```

```python
sns.barplot(y=match_wins.index,x=match_wins.values,palette='Set2')
plt.title("Most Matches wins by team")
```

```python
sns.countplot(x=df['toss_decision'],palette='Set2')
plt.title("Toss Decision")
```

# Toss winner vs Match winner

```python
count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=(count*100)/df.shape[0]
percentage.round(2)
```

# Runs winner vs Wicket winner

```python
sns.countplot(x=df['won_by'],palette='Set2')
plt.title("Won By")
```

# Key player performance 

```python
count=df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x=count.values,y=count.index,palette='Set2')
plt.title("Top 10 Player of the IPL")
```

```python
high=df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind='barh')
```

```python
df['highest_wickets']=df['best_bowling_figure'].apply(lambda x  :x.split('-')[0])
df['highest_wickets']=df['highest_wickets'].astype(int)
df
```

```python
top_bowlers=df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers
```

```python
top_bowlers.plot(kind='barh')
```

```python
vanue_count=df['venue'].value_counts().head(10)
vanue_count
```

```python
sns.barplot(y = vanue_count.index, x = vanue_count.values,palette='rainbow')
```
